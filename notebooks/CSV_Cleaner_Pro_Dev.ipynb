```python
# Step A: inspect, clean only non-numeric-like sales rows, and show before/after for them

# 1) Recreate the mask you used earlier (rows where sales string is not purely digits/dot/comma/minus)
mask_non_numeric = ~df['sales'].astype(str).str.match(r'^[\d\.\-,]+$')

print("Rows flagged as non-numeric (before cleaning):", mask_non_numeric.sum())
display(df.loc[mask_non_numeric, ['order_id','customer_name','unit_price','sales']])

# 2) For only these rows, remove any non-digit/non-dot/non-minus characters (e.g., 'USD ', '₹', commas)
df.loc[mask_non_numeric, 'sales'] = (
    df.loc[mask_non_numeric, 'sales']
      .astype(str)
      .str.replace(r'[^0-9\.\-]', '', regex=True)
)

# 3) Convert those cleaned strings to numeric (float) in-place
df.loc[mask_non_numeric, 'sales'] = pd.to_numeric(df.loc[mask_non_numeric, 'sales'], errors='coerce')

# 4) Show the same rows after cleaning
print("\nAfter cleaning (should be numeric):")
display(df.loc[mask_non_numeric, ['order_id','customer_name','unit_price','sales']])
print("\nDtype of sales now:", df['sales'].dtype)
```

```python
# Step B: convert the entire 'sales' column to float safely
df['sales'] = pd.to_numeric(df['sales'], errors='coerce')

# Verify conversion
print(df['sales'].dtype)
df[['quantity', 'unit_price', 'sales']].sample(5)
```

```python
# Step C1: Compare actual vs computed sales
df['computed_sales'] = df['quantity'] * df['unit_price']
df['sales_diff'] = (df['sales'] - df['computed_sales']).abs()

# Show rows where the difference is noticeable (> 0.5)
mismatch = df[df['sales_diff'] > 0.5][['order_id','quantity','unit_price','sales','computed_sales','sales_diff']]
print("Mismatched rows count:", len(mismatch))
display(mismatch.head(10))
```

```python
# Step C3: Recompute and replace sales for accuracy
df.loc[df['sales_diff'] > 0.5, 'sales'] = df['computed_sales']
print("✅ Recomputed sales for mismatched rows.")
```

```python
print("Remaining mismatches:", (df['sales'] - df['computed_sales']).abs().gt(0.5).sum())
```
